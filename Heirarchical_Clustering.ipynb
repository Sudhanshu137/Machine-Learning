```python
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Distance matrix
dist_matrix = np.array([
    [0, 9, 3, 6, 11],
    [9, 0, 7, 5, 10],
    [3, 7, 0, 9, 2],
    [6, 5, 9, 0, 8],
    [11, 10, 2, 8, 0]
])

# Convert the distance matrix into a condensed form required for linkage
condensed_dist_matrix = sch.distance.squareform(dist_matrix)

# Perform hierarchical clustering using the 'complete' method (as DIANA is not directly available in scipy)
linkage_matrix = sch.linkage(condensed_dist_matrix, method='complete')

# Plot dendrogram
plt.figure(figsize=(8, 5))
sch.dendrogram(linkage_matrix, labels=["A", "B", "C", "D", "E"])
plt.title("DIANA (Divisive Analysis Clustering) Approximation")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```
